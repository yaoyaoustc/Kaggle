# load data

## data description
Elevation - Elevation in meters
Aspect - Aspect in degrees azimuth
Slope - Slope in degrees
Horizontal_Distance_To_Hydrology - Horz Dist to nearest surface water features
Vertical_Distance_To_Hydrology - Vert Dist to nearest surface water features
Horizontal_Distance_To_Roadways - Horz Dist to nearest roadway
Hillshade_9am (0 to 255 index) - Hillshade index at 9am, summer solstice
Hillshade_Noon (0 to 255 index) - Hillshade index at noon, summer solstice
Hillshade_3pm (0 to 255 index) - Hillshade index at 3pm, summer solstice
Horizontal_Distance_To_Fire_Points - Horz Dist to nearest wildfire ignition points
Wilderness_Area (4 binary columns, 0 = absence or 1 = presence) - Wilderness area designation
Soil_Type (40 binary columns, 0 = absence or 1 = presence) - Soil Type designation
Cover_Type (7 types, integers 1 to 7) - Forest Cover Type designation

```python
def num_cat_sep(df_X):
    droplist = df_X.columns[df_X.columns.str.contains('Soil_Type')]
    drop = droplist.append(df_X.columns[df_X.columns.str.contains('Wilderness_Area')])
    X_numerical = df_X.drop(columns=drop,axis=1)
    X_categorical = df_X.loc[:,drop]
    return X_numerical,X_categorical
```

```python
def FeatureEngineering(df_train,df_test):
    # seperate x y for df_train
    df_train_X = df_train.drop(columns='Cover_Type')
    df_train_Y = df_train.loc[:,'Cover_Type']
    # mark trainset
    df_train_X.loc[:,'trainset'] = 1
    df_test.loc[:,'trainset'] = 0
    df_concat_X = pd.concat([df_train_X,df_test])
    # drop columns
    dropcol = ['Id','Wilderness_Area2']
    df_concat_X.drop(columns=dropcol,inplace = True)
    # seperate numerical
    df_concat_X_num,df_concat_X_cat = num_cat_sep(df_concat_X)
    # fix skew dataset
    df_concat_X_num = skewfix(df_concat_X_num)
    # combine num and cat
    df_concat_X = pd.concat([df_concat_X_num,df_concat_X_cat],axis=1)
    
    # split train and test
    X_train = df_concat_X[df_concat_X.loc[:,'trainset'] == 1].drop(columns = 'trainset')
    X_test = df_concat_X[df_concat_X.loc[:,'trainset'] == 0].drop(columns = 'trainset')
    return X_train, X_test, df_train_Y
```

```python
X_train,X_test,Y_train = FeatureEngineering(df_train,df_test)
```
